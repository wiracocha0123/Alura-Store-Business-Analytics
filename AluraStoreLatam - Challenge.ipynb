````python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

df_total = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

df_total.head()
````

## Limpieza y Preprocesamiento de Datos

### Subtask:
Estandarizar los nombres de las columnas para facilitar el análisis (si es necesario) y convertir las columnas 'Precio', 'Costo de envío' y 'Calificación' a tipos de datos numéricos adecuados, manejando cualquier valor faltante o inconsistente.


**Reasoning**:
First, I'll standardize the column names in `df_total` to ensure consistency and ease of access. This involves converting them to lowercase and replacing spaces with underscores.



````python
df_total.columns = df_total.columns.str.lower().str.replace(' ', '_')
print("Nombres de columnas estandarizados:\n", df_total.columns)
````

**Reasoning**:
Next, I will convert the 'precio', 'costo_de_envío', and 'calificación' columns to numeric types, coercing any non-numeric values to NaN, and then display the data types and the first few rows of the DataFrame to verify the conversion and inspect for NaNs.



````python
df_total['precio'] = pd.to_numeric(df_total['precio'], errors='coerce')
df_total['costo_de_envío'] = pd.to_numeric(df_total['costo_de_envío'], errors='coerce')
df_total['calificación'] = pd.to_numeric(df_total['calificación'], errors='coerce')

print("Tipos de datos de las columnas relevantes después de la conversión:\n", df_total[['precio', 'costo_de_envío', 'calificación']].dtypes)
print("\nPrimeras filas del DataFrame después de la conversión:\n")
df_total.head()
````

**Reasoning**:
Now that the columns have been converted to numeric types, I will check for any NaN values that might have been introduced during the conversion process in the 'precio', 'costo_de_envío', and 'calificación' columns. This is crucial for data quality and to decide on an appropriate handling strategy for these missing values.



````python
print("Valores NaN después de la conversión:")
print(df_total[['precio', 'costo_de_envío', 'calificación']].isnull().sum())
````

## Calcular Ingreso Total por Tienda

### Subtask:
Calcular el ingreso total de cada tienda sumando los valores de la columna 'precio' para cada una de ellas.


**Reasoning**:
The subtask requires calculating the total revenue for each store. This can be achieved by grouping the `df_total` DataFrame by the 'tienda' column and then summing the 'precio' column for each group. The result will be stored in `ingreso_total_por_tienda` and then printed.



````python
ingreso_total_por_tienda = df_total.groupby('tienda')['precio'].sum()
print("Ingreso total por tienda:\n", ingreso_total_por_tienda)
````

## Analizar Ventas por Categoría de Producto

### Subtask:
Agrupar los datos por tienda y 'Categoría del Producto' y contar el número de ventas para cada categoría, identificando las categorías más populares en cada tienda.


**Reasoning**:
To analyze sales by category, I will group the `df_total` DataFrame by 'tienda' and 'categoría_del_producto' and count the number of sales using `.size()`. This will provide a series with the number of sales for each category within each store.



````python
ventas_por_categoria = df_total.groupby(['tienda', 'categoría_del_producto']).size().reset_index(name='numero_de_ventas')
print("Número de ventas por categoría y tienda:")
print(ventas_por_categoria.head())

print("\nCategorías más populares por tienda (primeras 5 de cada tienda, ordenadas por número de ventas):")
for tienda_name in ventas_por_categoria['tienda'].unique():
    print(f"\n--- {tienda_name} ---")
    print(ventas_por_categoria[ventas_por_categoria['tienda'] == tienda_name].sort_values(by='numero_de_ventas', ascending=False).head())
````

## Calcular Calificación Promedio por Tienda

### Subtask:
Calcular la calificación promedio de los clientes para cada tienda utilizando la columna 'Calificación' para evaluar la satisfacción del cliente.


**Reasoning**:
To calculate the average rating per store, I will group the `df_total` DataFrame by the 'tienda' column and then compute the mean of the 'calificación' column for each group. The result will be stored in `calificacion_promedio_por_tienda` and then printed to display the average ratings.



````python
calificacion_promedio_por_tienda = df_total.groupby('tienda')['calificación'].mean()
print("Calificación promedio por tienda:\n", calificacion_promedio_por_tienda)
````

## Identificar Productos Más y Menos Vendidos

### Subtask:
Contar las ocurrencias de cada 'Producto' dentro de cada tienda para identificar los productos más vendidos y los menos vendidos.


**Reasoning**:
To identify the most and least sold products per store, I will first group the `df_total` DataFrame by 'tienda' and 'producto', and then count the occurrences of each product. After that, I'll display the top 5 most sold and least sold products for each store.



````python
ventas_por_producto = df_total.groupby(['tienda', 'producto']).size().reset_index(name='conteo_ventas')

print("Ventas por producto y tienda (primeras filas):\n", ventas_por_producto.head())

print("\n--- Productos más y menos vendidos por tienda ---")
for tienda_name in ventas_por_producto['tienda'].unique():
    print(f"\nTienda: {tienda_name}")

    # Productos más vendidos
    mas_vendidos = ventas_por_producto[ventas_por_producto['tienda'] == tienda_name].sort_values(by='conteo_ventas', ascending=False).head(5)
    print("Más vendidos:\n", mas_vendidos)

    # Productos menos vendidos
    menos_vendidos = ventas_por_producto[ventas_por_producto['tienda'] == tienda_name].sort_values(by='conteo_ventas', ascending=True).head(5)
    print("Menos vendidos:\n", menos_vendidos)
````

## Calcular Costo de Envío Promedio por Tienda

### Subtask:
Calcular el costo de envío promedio para cada tienda utilizando la columna 'Costo de envío'.


**Reasoning**:
To calculate the average shipping cost per store, I will group the `df_total` DataFrame by the 'tienda' column and then compute the mean of the 'costo_de_envío' column for each group. The result will be stored in `costo_envio_promedio_por_tienda` and then printed to display the average shipping costs.



````python
costo_envio_promedio_por_tienda = df_total.groupby('tienda')['costo_de_envío'].mean()
print("Costo de envío promedio por tienda:\n", costo_envio_promedio_por_tienda)
````
